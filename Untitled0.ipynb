```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
from IPython.display import display, HTML

# Fonction du modèle de Monod modifié avec inhibition par le substrat
def modified_monod_model(S, Ks, Ki):
    mu_max = 1.0  # Taux de croissance spécifique maximal
    return (mu_max * S) / (Ks + S + (S**2 / Ki))

# Fonction pour mettre à jour le graphique
def update_plot(Ks, Ki):
    S = np.linspace(0, 10, 100)  # Plage de concentration de substrat
    mu = modified_monod_model(S, Ks, Ki)

    plt.figure(figsize=(8, 6))
    plt.plot(S, mu, label=f'Ks = {Ks}, Ki = {Ki}')
    plt.xlabel('Concentration de substrat (S)')
    plt.ylabel('Taux de croissance spécifique (mu)')
    plt.title('Simulation du modèle de Monod modifié avec inhibition par le substrat')
    plt.legend()
    plt.grid(True)
    plt.show()

# Curseurs interactifs pour Ks et Ki
Ks_slider = FloatSlider(value=0.5, min=0.1, max=5.0, step=0.1, description='Ks:')
Ki_slider = FloatSlider(value=1.0, min=0.1, max=100.0, step=1.0, description='Ki:')

# Graphique interactif
interact(update_plot, Ks=Ks_slider, Ki=Ki_slider)
```
